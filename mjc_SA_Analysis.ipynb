```python
def day_sum(row):
    openday=[]
    for i in range(len(row)):
            openday.append(float(row[i][1]))
    opensum=sum(openday)
    return float(opensum)


```

```python

def stock_return(column1, column2):
    stock=[0]*len(sap.yqdopen)
    for i in range(len(sap.yqdopen)):  
        daystock=[]
        for j in range(len(sap.yqdopen[i])):
            daystock.append((float(sap[str(column2)][i][j][1])-float(sap[str(column1)][i][j][1]))/float(sap[str(column2)][i][j][1]))
        stock[i]=daystock
    return pd.Series(stock)
```

```python
def count_unique(mylist):
    uniques = list(set(mylist))
    outlist = []
    for unique in uniques:
        counter = 0
        for word in mylist:
            if word == unique:
                counter += 1
        outlist.append([unique, counter])
    return outlist
```
