```python
su = pd.read_csv('/Users/michaelchuang/git_project/mjc_WebScraping/webscraping/lightning1.csv')
```

```python
su['bearish'] = su.bearish.apply(lambda x: x[0:].split(','))
su['bullish'] = su.bullish.apply(lambda x: x[0:].split(','))
```

```python
#remove exact same lists
j2=[]
for i in range(len(su.bullish)):
    if(list((set(su.bullish[i]) - set(su.bearish[i])))==[]):
        j2.append(i)
su.drop(su.index[j2])
```

```python
#remove duplicates in bullish
for i in range(len(su.bullish)):
    su.bullish[i]=list(set(su.bullish[i]) - set(su.bearish[i]))
```

```python
#remove non ticket links
for i in range(len(su.bullish)):
     su.bullish[i]= [x for x in su.bullish[i] if x.startswith('https://seekingalpha.com/symbol/')]        
```

```python
#remove seeking alpha
 
for i in range(len(su.bullish)):
      su.bullish[i]= [x.lstrip('https://seekingalpha.com/symbol/') for x in su.bullish[i]]
```

```python
for i in range(len(su.bearish)):
     su.bearish[i]= [x for x in su.bearish[i] if x.startswith('https://seekingalpha.com/symbol/')]        
```

```python
import pandas as pd
import io
import requests
import time
 
def google_stocks(symbol, startdate = (1, 1, 2005), enddate = None):
 
    startdate = str(startdate[0]) + '+' + str(startdate[1]) + '%2C+' + str(startdate[2])
 
    if not enddate:
        enddate = time.strftime("%m+%d+%Y")
    else:
        enddate = str(enddate[0]) + '+' + str(enddate[1]) + '%2C+' + str(enddate[2])
    stock_url='http://www.google.com/finance/historical?q=GOOG&startdate=Jul%2008%2C%202017&enddate=Aug%2008%2C%202017&output=csv'
    #stock_url = "http://www.google.com/finance/historical?q=" + symbol + \
    #            "&startdate=" + startdate + "&enddate=" + enddate + "&num=14"+"&output=csv"
    print(stock_url)
    raw_response = requests.get(stock_url).content
 
    stock_data = pd.read_csv(io.StringIO(raw_response.decode('utf-8')))
 
    return stock_data
```
