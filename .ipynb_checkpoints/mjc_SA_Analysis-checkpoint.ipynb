```python
sap = pd.read_csv('/Users/michaelchuang/git_project/mjc_WebScraping/webscraping/results3.csv', index_col='Unnamed: 0')
sap=sap[sap['yqdopen'] != '[]']
sap=sap.reset_index()
del sap['index']
# de-string prices
sap['yqdopen']=sap['yqdopen'].apply(lambda x: ast.literal_eval(x))
sap['yqdclose6m']=sap['yqdclose6m'].apply(lambda x: ast.literal_eval(x))
sap['yqdclose30d']=sap['yqdclose30d'].apply(lambda x: ast.literal_eval(x))
sap['yqdclose7d']=sap['yqdclose7d'].apply(lambda x: ast.literal_eval(x))
sap['bullish']=sap['bullish'].apply(lambda x: ast.literal_eval(x))
sap['bearish']=sap['bearish'].apply(lambda x: ast.literal_eval(x))


# Set to datetime and sort
sap['date']=pd.to_datetime(sap['date'], format='%Y%m%d')
sap['enddate']=pd.to_datetime(sap['enddate'], format='%Y%m%d')
sap=sap.sort_values(by='date', ascending=False)

#fixing missing outliers
sap.iloc[634].yqdclose6m[1][1]= '66.58'
sap.iloc[1080].yqdopen[2][1]='25.0'
sap.iloc[431][7][4][1]='0.195000'
```

```python
#function to add prices for a particular day

def day_sum(row):
    openday=[]
    for i in range(len(row)):
            openday.append(float(row[i][1]))
    opensum=sum(openday)
    return float(opensum)


```

```python
#function to return stock return between 2 columns, returns are in a  nested list
def stock_return(column1, column2):
    stock=[0]*len(sap.yqdopen)
    for i in range(len(sap.yqdopen)):  
        daystock=[]
        for j in range(len(sap.yqdopen[i])):
            daystock.append((float(sap[str(column2)][i][j][1])-float(sap[str(column1)][i][j][1]))/float(sap[str(column1)][i][j][1]))
        stock[i]=daystock
    return pd.Series(stock)
```

```python
#Adding columns to calculate total return from a whole day 
sap['dayreturn7d']=sap.apply(lambda row: (day_sum(row['yqdclose7d']) - day_sum(row['yqdopen']))/(day_sum(row['yqdopen'])), axis = 1)
sap['dayreturn30d']=sap.apply(lambda row: (day_sum(row['yqdclose30d']) - day_sum(row['yqdopen']))/(day_sum(row['yqdopen'])), axis = 1)
sap['dayreturn6m']=sap.apply(lambda row: (day_sum(row['yqdclose6m']) - day_sum(row['yqdopen']))/(day_sum(row['yqdopen'])), axis = 1)

#Adding columns to calculate return from an individual stock
sap['indreturn7d'] = stock_return('yqdopen', 'yqdclose7d')
sap['indreturn30d'] = stock_return('yqdopen', 'yqdclose30d')
sap['indreturn6m'] = stock_return('yqdopen', 'yqdclose6m')
```

```python
#Adding column for day of the week for potential further analysis
sap['weekday'] = sap['date'].dt.dayofweek
```

```python
##Function to detemrine how many unique stocks were recommended
def count_unique(mylist):
    uniques = list(set(mylist))
    outlist = []
    for unique in uniques:
        counter = 0
        for word in mylist:
            if word == unique:
                counter += 1
        outlist.append([unique, counter])
    return outlist
```
